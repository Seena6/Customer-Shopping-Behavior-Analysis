```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column for age group
labels = ['Young Adult' , 'Adult' , 'Middle-aged','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column for purchase frequency days

frequency_mapping = {
    'Fortnightly' : 14,
    'Quarterly' : 90,
    'Annually' : 365,
    'Weekly' : 7,
    'Monthly' : 30,
    'Bi-Weekly' : 14,
    'Every 3 Months' : 90
}

df['purchase frequency days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase frequency days']].head(10)
```
